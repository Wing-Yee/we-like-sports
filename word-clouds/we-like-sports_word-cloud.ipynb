```python
#Read csvs for each event
df=pd.read_csv("/Users/admin/code/Wing-Yee/data-challenges/we-like-sports/Cleaned Data/sentiment_analysis_test.csv") 

#list of event names
event_names = df.source.unique()

#separate words for the event
words = df['tidy_tweets'].tolist()
words_sep = ''.join(words)


#method for getting black
def black_color_func(word, font_size, position,orientation,random_state=None, **kwargs):
    return("hsl(0,100%, 1%)")

# white background
# max_words to 500
# width and height 3000 x 2000
wordcloud = WordCloud(font_path = '/Library/Fonts/Arial Unicode.ttf', 
                      background_color="white", width=3000, height=2000, max_words=500).generate(words_sep)

# words to black
wordcloud.recolor(color_func = black_color_func)

# set the figsize
plt.figure(figsize=[15,10])

# plot the wordcloud
plt.imshow(wordcloud, interpolation="bilinear")

# remove plot axes
plt.axis("off")

# save the image
plt.savefig('cloud.jpg')
```
